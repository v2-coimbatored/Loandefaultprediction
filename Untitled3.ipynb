```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# Load the train dataset
train_data = pd.read_csv("/Users/viveksmac/Documents/UWE/Msc Fintech/Term 2/Machine Learning/train.csv")

# Define the predictor variables (features) and the target variable
predictor_variables = ['interest_rate', 'unpaid_principal_bal', 'loan_term', 'loan_to_value',
                       'debt_to_income_ratio', 'borrower_credit_score', 'insurance_percent',
                       'co-borrower_credit_score', 'm1', 'm2', 'm3', 'm4', 'm5', 'm6', 'm7', 'm8', 'm9', 'm10', 'm11', 'm12']
target_variable = 'm13'

# Preprocessing for train data
# Impute missing numerical values with mean
imputer_num = SimpleImputer(strategy='mean')
train_data[predictor_variables] = imputer_num.fit_transform(train_data[predictor_variables])

# Impute missing categorical values with most frequent value
categorical_columns = ['source', 'financial_institution', 'loan_purpose', 'insurance_type']
for column in categorical_columns:
    train_data[column].fillna(train_data[column].mode()[0], inplace=True)

# Scale numerical features
scaler = StandardScaler()
train_data[predictor_variables] = scaler.fit_transform(train_data[predictor_variables])

# Split the dataset into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_data[predictor_variables], train_data[target_variable], test_size=0.2, random_state=42)

# Ensure DataFrames are used for training
X_train = pd.DataFrame(X_train, columns=predictor_variables)
X_test = pd.DataFrame(X_test, columns=predictor_variables)

# Logistic Regression
logistic_model = LogisticRegression(max_iter=1000)  # Increase max_iter to ensure convergence
logistic_model.fit(X_train, y_train)

# Decision Tree
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(X_train, y_train)

# XGBoost
xgboost_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgboost_model.fit(X_train, y_train)

# LightGBM
lightgbm_model = LGBMClassifier()
lightgbm_model.fit(X_train, y_train)

# Load the test dataset
test_data = pd.read_csv("/Users/viveksmac/Documents/UWE/Msc Fintech/Term 2/Machine Learning/test.csv")

# Preprocessing for test data (ensure it matches the train data preprocessing)
# Impute missing numerical values with mean
test_data[predictor_variables] = imputer_num.transform(test_data[predictor_variables])

# Impute missing categorical values with most frequent value
for column in categorical_columns:
    test_data[column].fillna(test_data[column].mode()[0], inplace=True)

# Scale numerical features
test_data[predictor_variables] = scaler.transform(test_data[predictor_variables])

# Ensure DataFrames are used for testing
X_test_final = pd.DataFrame(test_data[predictor_variables], columns=predictor_variables)

# Predict using the trained models
logistic_predictions = logistic_model.predict(X_test_final)
logistic_proba = logistic_model.predict_proba(X_test_final)[:, 1]

decision_tree_predictions = decision_tree_model.predict(X_test_final)
decision_tree_proba = decision_tree_model.predict_proba(X_test_final)[:, 1]

xgboost_predictions = xgboost_model.predict(X_test_final)
xgboost_proba = xgboost_model.predict_proba(X_test_final)[:, 1]

lightgbm_predictions = lightgbm_model.predict(X_test_final)
lightgbm_proba = lightgbm_model.predict_proba(X_test_final)[:, 1]

# Assuming the target variable is in the test dataset for evaluation
if 'm13' in test_data.columns:
    y_test_final = test_data['m13']

    # Evaluate the models
    logistic_accuracy = accuracy_score(y_test_final, logistic_predictions)
    logistic_auc = roc_auc_score(y_test_final, logistic_proba)

    decision_tree_accuracy = accuracy_score(y_test_final, decision_tree_predictions)
    decision_tree_auc = roc_auc_score(y_test_final, decision_tree_proba)

    xgboost_accuracy = accuracy_score(y_test_final, xgboost_predictions)
    xgboost_auc = roc_auc_score(y_test_final, xgboost_proba)

    lightgbm_accuracy = accuracy_score(y_test_final, lightgbm_predictions)
    lightgbm_auc = roc_auc_score(y_test_final, lightgbm_proba)

    print("Logistic Regression: Accuracy = {}, AUC = {}".format(logistic_accuracy, logistic_auc))
    print("Decision Tree: Accuracy = {}, AUC = {}".format(decision_tree_accuracy, decision_tree_auc))
    print("XGBoost: Accuracy = {}, AUC = {}".format(xgboost_accuracy, xgboost_auc))
    print("LightGBM: Accuracy = {}, AUC = {}".format(lightgbm_accuracy, lightgbm_auc))
else:
    print("The test dataset does not contain the target variable for evaluation.")
```
